### 1.1 Rate order berdasarkan waktu
----

```python
# Query that returns the number of products sold by times
# Then, store to dataframe
query_order_time = """
SELECT
	order_purchase_timestamp,
	COUNT(order_id) AS order_count
FROM
	olist_order_dataset
GROUP BY
	order_purchase_timestamp
ORDER BY
	order_purchase_timestamp
"""
df_order_time = pd.read_sql_query(query_order_time, conn)
df_order_time.head(10)
```

```python
# Query that returns the order_purchase_timestamp and the product_category_name
# and the order_id product_id from olist_order_items_dataset
# and the product_category_name from olist_products_dataset
# Then, store to dataframe
query_order_time_product = """
SELECT
	olist_order_dataset.order_purchase_timestamp,
	olist_order_items_dataset.product_id,
	olist_order_items_dataset.order_id,
	olist_products_dataset.product_category_name
FROM
	olist_order_dataset
INNER JOIN
	olist_order_items_dataset
ON
	olist_order_dataset.order_id = olist_order_items_dataset.order_id
INNER JOIN
	olist_products_dataset
ON
	olist_order_items_dataset.product_id = olist_products_dataset.product_id
GROUP BY
	olist_order_dataset.order_purchase_timestamp
ORDER BY
	olist_order_dataset.order_purchase_timestamp
"""
df_order_time_product = pd.read_sql_query(query_order_time_product, conn)
df_order_time_product.head()
```

### 1.2 Kategori produk yang paling banyak terjual beserta reviewnya berdasarkan waktu order

```python
# Query that returns the order_purchase_timestamp from order_id from olist_order_items_dataset
# then review_score and review_comment_message from olist_order_reviews_dataset
# Group by order_purchase_timestamp, Order by order_purchase_timestamp
# Then, store to dataframe
query_order_time_review = """
SELECT
	olist_order_dataset.order_purchase_timestamp,
	olist_order_items_dataset.order_id,
	olist_order_reviews_dataset.review_score,
	olist_order_reviews_dataset.review_comment_message
FROM
	olist_order_dataset
INNER JOIN
	olist_order_items_dataset
ON
	olist_order_dataset.order_id = olist_order_items_dataset.order_id
INNER JOIN
	olist_order_reviews_dataset
ON
	olist_order_items_dataset.order_id = olist_order_reviews_dataset.order_id
GROUP BY
	olist_order_dataset.order_purchase_timestamp
ORDER BY
	olist_order_dataset.order_purchase_timestamp
"""
df_order_time_review = pd.read_sql_query(query_order_time_review, conn)
df_order_time_review.head()
```

```python
# Query that returns number of order based on customer_id from olist_order_dataset
# Join with customer_city from olist_order_customer_dataset
# Then, store to dataframe
query_order_city = """
SELECT
	olist_order_dataset.customer_id,
	olist_order_customer_dataset.customer_city,
	COUNT(olist_order_dataset.order_id) AS order_count
FROM
	olist_order_dataset
INNER JOIN
	olist_order_customer_dataset
ON
	olist_order_dataset.customer_id = olist_order_customer_dataset.customer_id
GROUP BY
	olist_order_customer_dataset.customer_city
ORDER BY
	olist_order_customer_dataset.customer_city
"""

df_order_city = pd.read_sql_query(query_order_city, conn)
df_order_city.head()
```

```python
# Query that returns number of order based on customer_id from olist_order_dataset
# Join with their payment_type and payment_value from olist_order_payments_dataset
# Join with customer_city from olist_order_customer_dataset
# Then, store to dataframe
query_order_city_payment = """
SELECT
	olist_order_dataset.customer_id,
	olist_order_customer_dataset.customer_city,
	olist_order_payments_dataset.payment_type,
	olist_order_payments_dataset.payment_value
FROM
	olist_order_dataset
INNER JOIN
	olist_order_customer_dataset
ON
	olist_order_dataset.customer_id = olist_order_customer_dataset.customer_id
INNER JOIN
	olist_order_payments_dataset
ON
	olist_order_dataset.order_id = olist_order_payments_dataset.order_id
GROUP BY
	olist_order_customer_dataset.customer_city
ORDER BY
	olist_order_customer_dataset.customer_city
"""
df_order_city_payment = pd.read_sql_query(query_order_city_payment, conn)
df_order_city_payment.head()
```
